# GNSS Visibility & Link Budget Analysis

This notebook analyzes the GNSS visibility data.

```python
from pathlib import Path
import numpy as np
from orbit_determination.visibility_plots import VisibilityPlotter

# Define paths based on the project root
project_root = Path.cwd().parent
data_path = project_root / 'data'
vis_data_dir = data_path / 'processed' / 'visibility'

# Path to the reference trajectory data file 
trajectory_file = data_path / 'raw' / 'receiver_orbits' / 'eor' / 'sample1.txt'

# (Optional) Load True Anomaly 
try:
    # Assuming True Anomaly is in column 10 (index 9) - Adjust if necessary
    rcver_ta = np.loadtxt(trajectory_file, skiprows=1, usecols=(6,))
except Exception as e:
    print("Trajectory data not loaded. Outage CDF plot will be disabled.")
    rcver_ta = None
```

```python
# Initialize the Plotter
constellations = ["GPS", "Galileo", "GLONASS", "BDS"]
#constellations = ["GPS"]
save_dir = project_root / 'data' / 'processed' / 'plots'
save_dir.mkdir(parents=True, exist_ok=True)

plotter = VisibilityPlotter(
    data_path=vis_data_dir,
    constellations=constellations,
    rcver_ta=rcver_ta,
    plot_title="EOR early phase analysis"
)
```

```python
# 1. Plot C/N0 Histogram
plotter.plot_ctnr_distribution(min_ctnr=20, save_path=save_dir)

# 2. Plot C/N0 Sensitivity Curve
plotter.plot_ctnr_sensitivity(ref_ctnr=20, save_path=save_dir)

# 3. Plot Multi-GNSS Cumulative Visibility Time
plotter.plot_cumulative_visibility(save_path=save_dir)

# 4. Plot Outage Durations (Requires True Anomaly data)
if rcver_ta is not None:
    plotter.plot_outage_cdf(save_path=save_dir)
```
