# GNSS Constellation Generation & SGP4 Propagation

This notebook instantiates the `ConstellationSimulator` to parse Orbital Mean Elements (OMMs), filter non-operational satellites, and propagate reference trajectories for the GNSS constellations using the SGP4 algorithm.

```python
from pathlib import Path
from datetime import datetime
from orbit_determination.sgp4_propagation import ConstellationSimulator
```

## Simulation Configuration

```python
# Define the directory containing the OMM data
data_path = Path.cwd().parent / 'data/raw/omm'

# Initialize the simulator instance
simulator = ConstellationSimulator(data_path)

# Set the initial date of the simulation for data logging
dt_initial = datetime(2025, 3, 21, 5, 0, 0)
# Set simulation duration in seconds
sim_duration_s = 86400 * 4
# Set simulation time step in seconds
sim_step_s = 60
```

## Execution
Run the batch propagation. The simulator will automatically parse the data, transform the TEME coordinates to the GCRF (J2000) frame, split the BeiDou constellation by orbital regime, and export the `.npz` archives to the appropiate directory.

```python
# Execute the automated propagation pipeline
export_dir_path = Path.cwd().parent / 'data/interim/ephemeris'
simulator.run_all_constellations(dt_initial, sim_duration_s, sim_step_s, export_dir_path)
```
